**1. Count Odd and Even Numbers [Marks: 5]**

Using the provided integer.txt file, develop a spark script to count the number of odd and
even integers.

```python
# Read file and extract integer
numbers_rdd = spark.sparkContext.textFile('/FileStore/tables/integer.txt')

numbers_rdd = numbers_rdd.map(lambda row: int(row[0]))
numbers_rdd.collect()
```

```python
# Convert to int and map even/odd (0: even, 1: odd)
arrayRDD = numbers_rdd.map(lambda x: int(x) % 2)
arrayRDD.collect()
```

```python
# Map each (even/odd) into (key, 1)
reduceRDD = arrayRDD.map(lambda number: (number, 1))
reduceRDD.collect()
```

```python
# Reduce by key
result = reduceRDD.reduceByKey(lambda x, y: x + y)

#result
print(f"Even number: {result.collect()[0][1]}")
print(f"Odd number: {result.collect()[1][1]}")
```
